```python
monday = pd.Series({'Alice':8,'Bob':6,'David':7})
tuesday = pd.Series({'Bob':8,'Carol':5,'David':6})
wednesday = pd.Series({'Alice':7,'Carol':8,'David':5})
print(monday,tuesday,wednesday, sep='\n')
```

```python
tel_aviv = pd.Series({'milk': 10, 'bread': 5,  'eggs': 12, 'butter': 3})
haifa    = pd.Series({'milk': 8,  'eggs': 6,  'water': 20})
modiin   = pd.Series({'bread': 15, 'butter': 7, 'water': 10, 'juice': 5})
```

```python
total_stock =tel_aviv.add(haifa, fill_value=0).add(modiin, fill_value=0)

print(total_stock)
```

```python
product = input("Enter product name: ")

if product in total_stock:
    print(f"{product}: {int(total_stock[product])} units across all stores")
else:
    print(f"{product} is not stocked in any store")
    
```
